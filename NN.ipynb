```python
import numpy as np
import pandas as pd
```

```python
ds = pd.read_csv("train.csv")
```

```python
ds
```

```python
x = ds.iloc[:,[6,7,8,9,10]]
```

```python
x
```

```python
x["LoanAmount"].fillna(0, inplace = True) 
```

```python
x["Loan_Amount_Term"].fillna(0, inplace = True) 
```

```python
x["Credit_History"].fillna(0, inplace = True) 
```

```python
x
```

```python
y = ds.iloc[:,-1]
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder as le

y = le().fit_transform(y[0:614])
```

```python
from sklearn.preprocessing import StandardScaler as ss

x = ss().fit_transform(x)
```

```python
import keras
from keras.models import Sequential 
from keras.layers import Dense 
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 10, activation = "relu", input_dim = 5))
```

```python
model.add(Dense(units = 6, activation = "relu"))
```

```python
model.add(Dense(units = 4, activation = "relu"))
```

```python
model.add(Dense(units = 1, activation = "sigmoid"))
```

```python
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['acc'])
```

```python
model.fit(x, y, batch_size = 32, epochs = 100)
```
